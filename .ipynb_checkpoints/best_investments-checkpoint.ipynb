```python
import pandas as pd
import numpy as np
```

```python
stock = pd.read_csv("stock_price_clean.csv")
bond = pd.read_csv("GS1.csv")
housing = pd.read_csv("house_price_clean.csv")
```

```python
stock["date"] = pd.to_datetime(stock["Year_Month"], infer_datetime_format = True)
```

```python
stock["year"] = [x.year for x in stock["date"]]
```

```python
stock_mean = stock.groupby(['year', 'Company']).sum().sum(
    level=['year', 'Company']).unstack('year').fillna(0).reset_index()

# Fix the column names
# df.columns = ['Date', 'one', 'two']
```
